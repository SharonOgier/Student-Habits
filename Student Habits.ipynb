```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings

df = pd.read_csv("/kaggle/input/student-habits-vs-academic-performance/student_habits_performance.csv")

warnings.simplefilter(action='ignore', category=FutureWarning)


plt.figure(figsize=(8, 6))
sns.barplot(
    x='diet_quality',
    y='exam_score',
    data=df,
    estimator=np.mean,
    order=df.groupby('diet_quality')['exam_score'].mean().sort_values(ascending=False).index,
    palette='Blues_r'
)
plt.title('Average Exam Score by Diet Quality')
plt.xlabel('Diet Quality')
plt.ylabel('Average Exam Score')
plt.tight_layout()
plt.show()

# Generate additional visualizations
figures = []

# 1. Study hours per day vs Exam Score (scatter with regression line)
plt.figure(figsize=(8, 6))
sns.regplot(x='study_hours_per_day', y='exam_score', data=df, scatter_kws={'alpha':0.5}, line_kws={"color":"black"})
plt.title('Study Hours per Day vs Exam Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.tight_layout()
figures.append(plt.gcf())

# 2. Boxplot: Exam Score by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='exam_score', data=df, palette='Blues_r')
plt.title('Exam Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.tight_layout()
figures.append(plt.gcf())

# 3. Bar Chart: Average Exam Score by Internet Quality
plt.figure(figsize=(8, 6))
sns.barplot(
    x='internet_quality',
    y='exam_score',
    data=df,
    estimator=np.mean,
    order=df.groupby('internet_quality')['exam_score'].mean().sort_values(ascending=False).index,
    palette='Blues_r'
)
plt.title('Average Exam Score by Internet Quality')
plt.xlabel('Internet Quality')
plt.ylabel('Average Exam Score')
plt.tight_layout()
figures.append(plt.gcf())

# 4. Violin Plot: Sleep Hours by Exam Score Range (categorized)
df['exam_score_category'] = pd.cut(df['exam_score'], bins=[0, 50, 70, 100], labels=['Low', 'Medium', 'High'])
plt.figure(figsize=(8, 6))
sns.violinplot(x='exam_score_category', y='sleep_hours', data=df, palette='Blues_r')
plt.title('Sleep Hours Distribution by Exam Score Category')
plt.xlabel('Exam Score Category')
plt.ylabel('Sleep Hours')
plt.tight_layout()
figures.append(plt.gcf())

# Display all generated figures
for fig in figures:
    plt.figure(fig.number)
    plt.show()

# Generate more visualizations
more_figures = []

# 5. Countplot: Exercise Frequency (ordered)
plt.figure(figsize=(8, 6))
exercise_order = df['exercise_frequency'].value_counts().sort_values(ascending=False).index
sns.countplot(x='exercise_frequency', data=df, order=exercise_order, palette='Blues_r')
plt.title('Distribution of Exercise Frequency')
plt.xlabel('Exercise Frequency (Days/Week)')
plt.ylabel('Number of Students')
plt.tight_layout()
more_figures.append(plt.gcf())

# 6. Bar Chart: Average Exam Score by Part-Time Job
plt.figure(figsize=(8, 6))
sns.barplot(
    x='part_time_job',
    y='exam_score',
    data=df,
    estimator=np.mean,
    order=df.groupby('part_time_job')['exam_score'].mean().sort_values(ascending=False).index,
    palette='Blues_r'
)
plt.title('Average Exam Score by Part-Time Job Status')
plt.xlabel('Part-Time Job')
plt.ylabel('Average Exam Score')
plt.tight_layout()
more_figures.append(plt.gcf())

# 7. Bar Chart: Average Exam Score by Extracurricular Participation
plt.figure(figsize=(8, 6))
sns.barplot(
    x='extracurricular_participation',
    y='exam_score',
    data=df,
    estimator=np.mean,
    order=df.groupby('extracurricular_participation')['exam_score'].mean().sort_values(ascending=False).index,
    palette='Blues_r'
)
plt.title('Average Exam Score by Extracurricular Participation')
plt.xlabel('Extracurricular Participation')
plt.ylabel('Average Exam Score')
plt.tight_layout()
more_figures.append(plt.gcf())

# 8. Pairplot: Selected continuous variables with 'exam_score'

sns.pairplot(
    df,
    hue='gender',
    vars=['study_hours_per_day', 'social_media_hours', 'sleep_hours', 'exam_score'],
    palette='Set2'  # valid qualitative palette
)
plt.tight_layout()
plt.gcf().suptitle("Pairplot of Habits and Exam Score by Gender", y=1.02)
plt.show()


# Show other individual figures
for fig in more_figures:
    plt.figure(fig.number)
    plt.show()


# Generate further insightful visualizations
extra_figures = []

# 9. Bar Chart: Average Exam Score by Mental Health Rating
plt.figure(figsize=(8, 6))
sns.barplot(
    x='mental_health_rating',
    y='exam_score',
    data=df,
    estimator=np.mean,
    order=df.groupby('mental_health_rating')['exam_score'].mean().sort_values(ascending=False).index,
    palette='Blues_r'
)
plt.title('Average Exam Score by Mental Health Rating')
plt.xlabel('Mental Health Rating (1-10)')
plt.ylabel('Average Exam Score')
plt.tight_layout()
extra_figures.append(plt.gcf())

# 10. Boxplot: Exam Score by Parental Education Level
plt.figure(figsize=(10, 6))
order = df.groupby('parental_education_level')['exam_score'].mean().sort_values(ascending=False).index
sns.boxplot(
    x='parental_education_level',
    y='exam_score',
    data=df,
    order=order,
    palette='Blues_r'
)
plt.title('Exam Score by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Exam Score')
plt.xticks(rotation=30)
plt.tight_layout()
extra_figures.append(plt.gcf())

# 11. Histogram: Distribution of Attendance Percentage with KDE
plt.figure(figsize=(8, 6))
sns.histplot(df['attendance_percentage'], kde=True, bins=30, color=sns.color_palette('Blues_r')[4])
plt.title('Distribution of Attendance Percentage')
plt.xlabel('Attendance Percentage')
plt.ylabel('Frequency')
plt.tight_layout()
extra_figures.append(plt.gcf())

# Show all generated figures
for fig in extra_figures:
    plt.figure(fig.number)
    plt.show()
```
